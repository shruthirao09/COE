# converting the columns into appropriate datetype formats


```python
# data['OpeningDate'] = pd.to_datetime(data['OpeningDate'], errors='coerce')
# data['Deadline'] = pd.to_datetime(data['Deadline'], errors='coerce')
data['OpeningDate'] = pd.to_datetime(data['OpeningDate'], format="%d %B %Y, %I:%M %p", errors='coerce')
data['Deadline'] = pd.to_datetime(data['Deadline'], format="%d %B %Y, %I:%M %p", errors='coerce')
```

```python
data.columns = data.columns.str.strip()
```

# remove duplicates

```python
data = data.drop_duplicates()
```

# Convert numeric columns to proper types

```python
data.loc[:, "Year"] = pd.to_numeric(data["Year"], errors="coerce")
data.loc[:, "Month"] = pd.to_numeric(data["Month"], errors="coerce")
data.loc[:, "IT"] = data["IT"].astype(bool)
```

# Fill or drop missing values in key columns

```python
data.loc[:, "Salary"] = data["Salary"].fillna("Not Specified")
data.loc[:, "ApplicationP"] = data["ApplicationP"].fillna("Not Provided")
```

# Dropping columns with over 70% null values(after discussing with client)

```python
threshold = 0.3
cleaned = data.dropna(axis=1, thresh=int(threshold * len(data)))
print(cleaned.info())
```

#  Normalize "Full Time" Entries

```python
cleaned.loc[:,'Term']= cleaned['Term'].str.lower()
cleaned.loc[:,'Term']= cleaned['Term'].str.replace('-',' ')
cleaned.loc[:,'Term']= cleaned['Term'].str.strip()
```

```python
cleaned['Term'].value_counts()
```

# Mode Imputation on column 'Term' as it holds categorical data

```python
cleaned.loc[:, 'Term'] = cleaned['Term'].fillna(cleaned['Term'].mode()[0])
cleaned
```

```python
cleaned['Duration'].value_counts()
```

# removing typos, incosistent delimeters, case sensitivity in column "Duration"

```python
cleaned.loc[:,'Duration'] = (cleaned['Duration']
                       .str.lower()  
                       .str.replace('-', ' ')  
                       .str.replace(',', ' ')  
                       .str.replace(r'\s+', ' ', regex=True)  
                       .str.replace('.', ' ')
                       .str.replace('months', 'month')
                       .str.strip())

print(cleaned['Duration'].value_counts())
```

# Mode Imputation on column "Duration"

```python
mode_value = cleaned['Duration'].mode()[0]
cleaned.loc[:,'Duration'] = cleaned['Duration'].fillna(mode_value)
cleaned
```

```python
cleaned.loc[:, 'StartDate'] = cleaned['StartDate'].fillna(cleaned['OpeningDate'])#the openingDate values will be stored 
cleaned.loc[:, 'StartDate'] = cleaned['StartDate'].fillna(cleaned['Deadline'])#if the OpeningDate value is empty, it takes Deadline cause dealine gives the time frame
cleaned.loc[:, 'StartDate'] = cleaned['StartDate'].fillna(cleaned['StartDate'].mode()[0])#if niether is available, mode imputation is implimented
```

```python
cleaned.loc[:, 'Deadline'] = cleaned.groupby(['Title', 'Company'])['Deadline'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else x.mean()))
cleaned.loc[:, 'Deadline'] = cleaned['Deadline'].fillna(cleaned['Deadline'].median())
cleaned.loc[:, 'Deadline'] = cleaned['Deadline'].fillna(cleaned['OpeningDate'] + pd.Timedelta(days=30))
cleaned.loc[:, 'Deadline'] = cleaned['Deadline'].fillna(pd.Timestamp('2000-01-01'))
cleaned.isnull().sum()
```

```python
import numpy as np
# 1. Fill using mode within Title & Company
cleaned.loc[:, 'OpeningDate'] = cleaned.groupby(['Title', 'Company'])['OpeningDate'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan))

# 2. Fill using median OpeningDate
cleaned.loc[:, 'OpeningDate'] = cleaned['OpeningDate'].fillna(cleaned['OpeningDate'].median())

# 3. Fill using Year & Month info (assuming mid-month)
cleaned.loc[:, 'OpeningDate'] = cleaned['OpeningDate'].fillna(pd.to_datetime(cleaned['Year'].astype(str) + '-' + cleaned['Month'].astype(str) + '-15'))

# 4. Assign a default placeholder date
cleaned.loc[:, 'OpeningDate'] = cleaned['OpeningDate'].fillna(pd.Timestamp('2000-01-01'))
cleaned.isnull().sum()
```

```python
import pandas as pd
import numpy as np
import warnings

# Suppress future warnings if you prefer (not recommended long term)
warnings.simplefilter(action='ignore', category=FutureWarning)

# Ensure 'RequiredQual' is treated as a string type
cleaned['RequiredQual'] = cleaned['RequiredQual'].astype(str)

# 1. Fill using mode within Title & Company
cleaned['RequiredQual'] = cleaned.groupby(['Title', 'Company'])['RequiredQual'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)

# 2. Fill using the most frequent qualification value within Jobpost's category
cleaned['RequiredQual'] = cleaned.groupby(['jobpost'])['RequiredQual'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)

# 3. Fill missing `RequiredQual` values with the most frequent qualification from the `Title`
cleaned['RequiredQual'] = cleaned.groupby('Title')['RequiredQual'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else np.nan)
)

# 4. Fallback - if all filling methods still leave nulls, use 'Not specified'
cleaned['RequiredQual'] = cleaned['RequiredQual'].fillna('Not specified')

# Ensure column types are inferred correctly after the fill
cleaned = cleaned.infer_objects()

# Check the missing values after filling
cleaned.isnull().sum()
```
